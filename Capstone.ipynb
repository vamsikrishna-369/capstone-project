```python
x=df[["Years_at_Company","Satisfaction_Level","Average_Monthly_Hours"]]
y=df["Salary"]
```

```python
!pip install scikit-learn
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
x
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(x_train)
```

```python
X_train
```

```python
import joblib
joblib.dump(scaler,"scaler.pkl")
```

```python
X_test=scaler.transform(x_test)
```

```python
x_test
```

```python
import numpy as np
from sklearn.metrics import mean_squared_error , mean_absolute_error
def results(prediction):
  print(f"Mean absolute error on model is {mean_absolute_error(y_test,prediction)}")
  print(f"Mean squared error on model is {np.sqrt(mean_squared_error(y_test,prediction))}")
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(X_train,y_train)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(x_train)
X_test = scaler.transform(x_test)
```

```python
predictions = lr.predict(X_test)
results(predictions)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVR
svrmodel = SVR()
param_gridsvr={
    'C':[0.1,1,10],
    'epsilon':[0.1,0.2,0.5],
    'kernel':['linear','poly','rbf'],
    'degree':[2,3,4]
}
gridsvr = GridSearchCV(svrmodel,param_gridsvr)
gridsvr.fit(X_train,y_train.values.ravel())
```

```python
gridsvr.best_params_
```

```python
predictionssvr = gridsvr.predict(x_test)
```

```python
results(predictionssvr)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
rfmodel = RandomForestRegressor()
param_gridrfr={
    'n_estimators':[2,3,4,5,6],
    'max_depth':[5,10,15]
}
gridrfr = GridSearchCV(rfmodel,param_gridrfr)
gridrfr.fit(x_train,y_train.values.ravel())
```

```python
gridrfr.best_params_

```

```python
predictionofrfr = gridrfr.predict(x_test)

```
